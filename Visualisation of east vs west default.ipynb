```python
import matplotlib.pyplot as plt
import numpy as np

# Variables
vars_main = ["10u", "10v", "2t", "t", "u", "v", "q"]
vars_heavy = ["msl", "z"]

# East values
east_rmse = [4.1901, 4.1474, 4.6401, 5.1504, 11.1195, 10.1203, 0.0009]
east_acc =  [0.0232, 0.1486, 0.8248, 0.9832, 0.5549, 0.3195, 0.9083]
east_accuracy = [0.5173, 0.4832, 0.3526, 0.3545, 0.2326, 0.1944, 1.0]

# West values
west_rmse = [4.2618, 3.6790, 6.0031, 4.4878, 10.9319, 11.3951, 0.0014]
west_acc =  [0.2785, 0.2859, 0.4198, 0.9906, 0.4054, 0.2030, 0.8597]
west_accuracy = [0.4759, 0.5038, 0.3347, 0.4091, 0.2052, 0.2119, 1.0]

# MSL and Z
east_msl_rmse, east_z_rmse = 1024.6957, 1058.0941
west_msl_rmse, west_z_rmse = 779.2924, 873.0838
east_msl_acc, east_z_acc = -0.0773, 0.9999
west_msl_acc, west_z_acc = 0.2825, 0.9999
east_msl_accuracy, east_z_accuracy = 0.0021, 0.0022
west_msl_accuracy, west_z_accuracy = 0.0013, 0.0017

x = np.arange(len(vars_main))
width = 0.35

# Main variable plots
def plot_comparison(title, east_vals, west_vals, ylabel):
    fig, ax = plt.subplots(figsize=(10, 5))
    ax.bar(x - width/2, east_vals, width, label='East', color='blue')
    ax.bar(x + width/2, west_vals, width, label='West', color='red')
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    ax.set_xticks(x)
    ax.set_xticklabels(vars_main)
    ax.legend()
    ax.grid(True, axis='y', linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()

plot_comparison("RMSE Comparison (excluding msl & z)", east_rmse, west_rmse, "RMSE")
plot_comparison("ACC Comparison (excluding msl & z)", east_acc, west_acc, "Anomaly Correlation Coefficient")
plot_comparison("Accuracy Comparison (excluding msl & z)", east_accuracy, west_accuracy, "Accuracy")

# Separate MSL and Z plots
def plot_heavy_vars(title, east_vals, west_vals, ylabel):
    fig, ax = plt.subplots(figsize=(6, 4))
    x_pos = np.arange(len(vars_heavy))
    ax.bar(x_pos - width/2, east_vals, width, label='East', color='blue')
    ax.bar(x_pos + width/2, west_vals, width, label='West', color='red')
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    ax.set_xticks(x_pos)
    ax.set_xticklabels(vars_heavy)
    ax.legend()
    ax.grid(True, axis='y', linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()

plot_heavy_vars("RMSE Comparison for msl & z", [east_msl_rmse, east_z_rmse], [west_msl_rmse, west_z_rmse], "RMSE")
plot_heavy_vars("ACC Comparison for msl & z", [east_msl_acc, east_z_acc], [west_msl_acc, west_z_acc], "Anomaly Correlation Coefficient")
plot_heavy_vars("Accuracy Comparison for msl & z", [east_msl_accuracy, east_z_accuracy], [west_msl_accuracy, west_z_accuracy], "Accuracy")
```
